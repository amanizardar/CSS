```python
tweets_list1 = []
for j in range(len(politician)):
    
    for i,tweet in enumerate(sntwitter.TwitterSearchScraper(f'from:{politician[j][1]} since:2022-03-10 until:2022-04-6').get_items()):
        # if i>100:
        #     break
        # print("hi")

        tweets_list1.append([tweet.user.username,tweet.user.id,tweet.user.description,tweet.user.verified,tweet.user.followersCount,tweet.user.friendsCount,tweet.user.listedCount,tweet.user.location,politician[j][0],tweet.content,tweet.date, tweet.id,tweet.url,tweet.replyCount,tweet.retweetCount,tweet.likeCount,tweet.quoteCount ])
    
    # politician_data[politician[j][1]]=tweets_list1
    # break
    # if(j==4):
    #     break
    
```

```python
len(tweets_list1)
```

```python
tweets_df1 = pd.DataFrame(tweets_list1, columns=['Username','Userid','Profile_Description','Verified?','Count_of_Followers','Count_of_Friends','Total_list_count','Location','Party','Tweet','Tweet_Datatime', 'Tweet_Id','Tweet_URL','Tweet_ReplyCount','Tweet_retweetCount','Tweet_LikeCount','Tweet_quoteCount'])
```
